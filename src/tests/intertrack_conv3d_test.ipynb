```python
import torch
import torch.nn as nn

batch_size = 16
in_channels = 2
width = 128
height = 128
track_num = 5
input_tensor = torch.randn([batch_size, track_num, in_channels, width, height])

kernel_width = 3
kernel_height = 3

padding_width = (kernel_width - 1) // 2
padding_height = (kernel_height - 1) // 2

conv3d = nn.Conv3d(in_channels, in_channels*track_num, (track_num,kernel_width,kernel_height), stride=1, padding=(0,padding_width,padding_height))
conv3d.weight.data.fill_(0.)
conv3d.bias.data.fill_(0.)

input_3d = input_tensor.permute(0,2,1,3,4)
output_3d = conv3d(input_3d)
print(output_3d.shape)
output_3d = output_3d.reshape(batch_size, in_channels*track_num, width, height)

print(output_3d)
```

```python
assert torch.allclose(output_3d, torch.zeros_like(output_3d))

```

```python
for param in conv3d.parameters():
    print(param.requires_grad)
```
