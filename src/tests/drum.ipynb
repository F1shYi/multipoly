```python
import os
import muspy
import numpy as np
MIDI_FOLDER = "/root/autodl-tmp/multipoly/data/train_segs_filtered"
test_midi_fpaths = os.listdir(MIDI_FOLDER)
print(len(test_midi_fpaths))
```

```python
min_pitches = []
max_pitches = []
min_max_pairs = []
from tqdm import tqdm
for midi_fpath in tqdm(test_midi_fpaths):
    
    test_midi_fpath = os.path.join(MIDI_FOLDER, midi_fpath)
    music = muspy.read_midi(test_midi_fpath)
    for track in music.tracks:
        if track.is_drum:
            
    
            min_pitch = 128
            max_pitch = 0
            for note in track.notes:
                min_pitch = min(min_pitch, note.pitch)
                max_pitch = max(max_pitch, note.pitch) 
            min_pitches.append(min_pitch)
            max_pitches.append(max_pitch)

            break
        
```

```python

import matplotlib.pyplot as plt
from collections import Counter

def plot_data(data):
    # 统计不同元素的出现次数
    counter = Counter(data)
    most_common_element, most_common_count = counter.most_common(1)[0]  # 取出现次数最多的元素及其计数
    print(most_common_count, most_common_element)
    # 提取元素名称和对应的频次
    labels, values = zip(*counter.items())

    # 绘制直方图
    plt.bar(labels, values, color='skyblue', edgecolor='black')

    # 添加标题和标签
    plt.title("Element Frequency Histogram")
    plt.xlabel("Elements")
    plt.ylabel("Frequency")

    # 显示图像
    plt.show()
```

```python
plot_data(min_pitches)
print(min(min_pitches))
```

```python
plot_data(max_pitches)
print(max(max_pitches))
```
